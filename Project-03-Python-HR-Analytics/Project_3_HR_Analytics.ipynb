```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("‚úÖ Pandas version:", pd.__version__)
print("‚úÖ NumPy version:", np.__version__)
print("All libraries loaded successfully!")
```

```python
# Load HR dataset
df = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')

# Display basic info
print("‚úÖ Dataset loaded successfully!")
print(f"üìä Total rows: {len(df)}")
print(f"üìä Total columns: {len(df.columns)}")
print(f"\nüìã Columns:")
print(df.columns.tolist())
print("\nüîç First 5 rows:")
df.head()
```

```python
# Data exploration
print("üìä DATASET OVERVIEW")
print("="*50)

# Data types
print("\nüîπ Data Types:")
print(df.dtypes)

# Missing values check
print("\nüîπ Missing Values:")
print(df.isnull().sum())

# Basic statistics
print("\nüîπ Statistical Summary:")
df.describe()
```

```python
# Attrition Analysis
print("üìä ATTRITION ANALYSIS")
print("="*50)

# Count of employees who left vs stayed
attrition_counts = df['Attrition'].value_counts()
print("\nüîπ Attrition Count:")
print(attrition_counts)

# Percentage
attrition_pct = df['Attrition'].value_counts(normalize=True) * 100
print("\nüîπ Attrition Percentage:")
print(attrition_pct.round(2))

# Visualization
plt.figure(figsize=(8, 5))
attrition_counts.plot(kind='bar', color=['green', 'red'])
plt.title('Employee Attrition Distribution', fontsize=14, fontweight='bold')
plt.xlabel('Attrition')
plt.ylabel('Number of Employees')
plt.xticks(rotation=0)
plt.show()
```

```python
# Attrition by Department
print("üìä ATTRITION BY DEPARTMENT")
print("="*50)

# Crosstab
dept_attrition = pd.crosstab(df['Department'], df['Attrition'], normalize='index') * 100
print("\nüîπ Attrition Rate by Department (%):")
print(dept_attrition.round(2))

# Visualization
plt.figure(figsize=(10, 6))
dept_attrition['Yes'].plot(kind='bar', color='coral')
plt.title('Attrition Rate by Department', fontsize=14, fontweight='bold')
plt.xlabel('Department')
plt.ylabel('Attrition Rate (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Salary vs Attrition
print("üìä SALARY vs ATTRITION")
print("="*50)

# Average monthly income by attrition
salary_attrition = df.groupby('Attrition')['MonthlyIncome'].mean()
print("\nüîπ Average Monthly Income:")
print(salary_attrition.round(2))

# Visualization
plt.figure(figsize=(10, 6))
df.boxplot(column='MonthlyIncome', by='Attrition', figsize=(8, 6))
plt.title('Monthly Income Distribution by Attrition')
plt.suptitle('')  # Remove default title
plt.xlabel('Attrition')
plt.ylabel('Monthly Income ($)')
plt.tight_layout()
plt.show()
```

```python
# Age vs Attrition
print("üìä AGE vs ATTRITION")
print("="*50)

# Average age by attrition
age_attrition = df.groupby('Attrition')['Age'].mean()
print("\nüîπ Average Age:")
print(age_attrition.round(2))

# Age distribution
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
df[df['Attrition']=='No']['Age'].hist(bins=20, alpha=0.7, color='green', label='Stayed')
df[df['Attrition']=='Yes']['Age'].hist(bins=20, alpha=0.7, color='red', label='Left')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution by Attrition')
plt.legend()

plt.subplot(1, 2, 2)
df.boxplot(column='Age', by='Attrition')
plt.suptitle('')
plt.title('Age Boxplot by Attrition')

plt.tight_layout()
plt.show()
```

```python
# Years at Company vs Attrition
print("üìä YEARS AT COMPANY vs ATTRITION")
print("="*50)

# Average years
years_attrition = df.groupby('Attrition')['YearsAtCompany'].mean()
print("\nüîπ Average Years at Company:")
print(years_attrition.round(2))

# Visualization
plt.figure(figsize=(10, 6))
df.boxplot(column='YearsAtCompany', by='Attrition')
plt.suptitle('')
plt.title('Years at Company by Attrition', fontsize=14, fontweight='bold')
plt.xlabel('Attrition')
plt.ylabel('Years at Company')
plt.tight_layout()
plt.show()

# Years distribution
plt.figure(figsize=(10, 6))
years_dist = df.groupby(['YearsAtCompany', 'Attrition']).size().unstack(fill_value=0)
years_dist.plot(kind='bar', stacked=True, color=['green', 'red'], figsize=(12, 6))
plt.title('Attrition by Years at Company', fontsize=14, fontweight='bold')
plt.xlabel('Years at Company')
plt.ylabel('Number of Employees')
plt.legend(['Stayed', 'Left'])
plt.tight_layout()
plt.show()
```

```python
# Correlation Analysis
print("üìä CORRELATION ANALYSIS")
print("="*50)

# Convert Attrition to numeric (Yes=1, No=0)
df['Attrition_Numeric'] = df['Attrition'].map({'Yes': 1, 'No': 0})

# Select numeric columns
numeric_cols = ['Age', 'MonthlyIncome', 'YearsAtCompany', 'JobSatisfaction',
                'EnvironmentSatisfaction', 'WorkLifeBalance', 'Attrition_Numeric']

# Correlation matrix
corr_matrix = df[numeric_cols].corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Correlation Heatmap', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# Top correlations with Attrition
print("\nüîπ Top Correlations with Attrition:")
attrition_corr = corr_matrix['Attrition_Numeric'].sort_values(ascending=False)
print(attrition_corr)
```
